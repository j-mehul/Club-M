```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
fd=pd.read_csv('train.csv')
fd1=pd.read_csv('test.csv')
```

```python
df1=fd.copy()
test=fd1.copy()
```

```python
df1['Flag']=1
test['Flag']=0
```

```python
df1.shape , test.shape
```

```python
df1.head()
```

```python
df1['booking_date'] = pd.to_datetime(df1['booking_date'], format='%d-%m-%Y')
df1['checkin_date'] = pd.to_datetime(df1['checkin_date'], format='%d-%m-%Y')
df1['checkout_date'] = pd.to_datetime(df1['checkout_date'], format='%d-%m-%Y')
```

```python
test['booking_date'] = pd.to_datetime(test['booking_date'], format='%d-%m-%Y')
test['checkin_date'] = pd.to_datetime(test['checkin_date'], format='%d-%m-%Y')
test['checkout_date'] = pd.to_datetime(test['checkout_date'], format='%d-%m-%Y')
```

```python
df1=df1[df1['checkout_date']>'04-09-2014']
```

```python
df=df1.append(test,sort=False)
```

```python
df.shape
```

```python
df.head()
```

```python
df['month']=df['booking_date'].dt.month
df['year']=df['booking_date'].dt.year
df['week']=df['booking_date'].dt.week
df['weekday']=df['booking_date'].dt.weekday
```

```python
df['in_month']=df['checkin_date'].dt.month
#df['in_year']=df['checkin_date'].dt.year
df['in_week']=df['checkin_date'].dt.week
df['in_weekday']=df['checkin_date'].dt.weekday
```

```python
#df['out_month']=df['checkout_date'].dt.month
#df['out_year']=df['checkout_date'].dt.year
df['out_week']=df['checkout_date'].dt.week
df['out_weekday']=df['checkout_date'].dt.weekday
```

```python
df['Booking_days']=(df['checkout_date']  - df['checkin_date']).dt.days
df['Bookingbefore']=(df['checkin_date']  - df['booking_date']).dt.days
```

```python
df['Total']=df.numberofadults + df.numberofchildren                      
```

```python
df['Total_per']=df[['total_pax','Total']].max(axis=1)
```

```python
df['noofchild']=df['Total_per']-df['numberofadults']
```

```python
df['haschild']=np.where(df['noofchild']>0,1,0)
```

```python
df=df.drop(['numberofchildren','total_pax','Total'],1)
```

```python
dates=df.select_dtypes(include=['datetime64[ns]'])
```

```python
df=df.drop(dates,1)
```

```python
df.isna().sum()
```

```python
df['season_holidayed_code']=df['season_holidayed_code'].fillna(df['season_holidayed_code'].mode()[0])
df['state_code_residence']=df['state_code_residence'].fillna(df['state_code_residence'].mode()[0])                 
```

```python
lbe=['resort_id','memberid','reservationstatusid_code']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df[lbe]=df[lbe].apply(lambda col: le.fit_transform(col))
```

```python
df=df.drop('reservation_id',1)
```

```python
df=pd.get_dummies(df)
```

```python
X=df[df['Flag']==1]
test=df[df['Flag']==0]
```

```python
X.shape, test.shape
```

```python
train1=X.drop('amount_spent_per_room_night_scaled',1)
y=X['amount_spent_per_room_night_scaled']
```

```python
test=test.drop('amount_spent_per_room_night_scaled',1)
```

```python
from sklearn.model_selection import train_test_split
train_x, val_x, train_y, val_y= train_test_split(train1,y, test_size=0.20, random_state=2019)
```

```python
train_x.shape , val_x.shape , test.shape
```

```python
train_x['amount_spent_per_room_night_scaled'] = train_y
```

```python
col=['resort_region_code','resort_type_code','room_type_booked_code',
     'state_code_residence','resort_id','memberid']
```

```python
col_dict=dict()
for col in col:
    p=train_x.groupby(col).agg({ 'amount_spent_per_room_night_scaled': [min, max,np.mean]}).reset_index()
    p.columns = [col +''.join(x) for x in p.columns.ravel()]
    col_dict[col] = p
```

```python
col_name='state_code_residence'
train_x=train_x.merge(col_dict[col_name],left_on=col_name,right_on=col_name+col_name,how='left')
val_x=val_x.merge(col_dict[col_name],left_on=col_name,right_on=col_name+col_name,how='left')
test=test.merge(col_dict[col_name],left_on=col_name,right_on=col_name+col_name,how='left')
```

```python
col_name='resort_id'
train_x=train_x.merge(col_dict[col_name],left_on=col_name,right_on=col_name+col_name,how='left')
val_x=val_x.merge(col_dict[col_name],left_on=col_name,right_on=col_name+col_name,how='left')
test=test.merge(col_dict[col_name],left_on=col_name,right_on=col_name+col_name,how='left')
```

```python
mm=train_x.groupby('memberid').state_code_residence.count().reset_index().sort_values('state_code_residence').rename(columns = {'state_code_residence':'a'})
```

```python
train_x=train_x.merge(mm,on = ['memberid'],how='left')
val_x=val_x.merge(mm,on = ['memberid'],how='left')
test=test.merge(mm,on = ['memberid'],how='left')
```

```python
import gc
gc.collect()
```

```python
train_x.info()
```

```python
train_x=train_x.drop(['resort_idresort_id','amount_spent_per_room_night_scaled',
                      'state_code_residencestate_code_residence','Flag'],1)
val_x=val_x.drop(['resort_idresort_id','state_code_residencestate_code_residence','Flag'],1)
test=test.drop(['resort_idresort_id','state_code_residencestate_code_residence','Flag'],1)
```

```python
val_x.head()
```

```python
import xgboost as xgb
feature=train_x.columns.tolist()
```

```python
param = {'objective' : 'reg:linear',
            'eta' : 0.01,
            'max_depth' : 10,
            'silent' : 1,
            'eval_metric' : "rmse",
            'min_child_weight' : 3,
            'nthread' : 16,
            'subsample' : 0.8,
            'colsample_bytree' : 0.8,
            'seed' : 1}

para = list(param.items())

xgtrain = xgb.DMatrix(train_x, train_y, feature_names=feature)
xgval = xgb.DMatrix(val_x, val_y, feature_names=feature)
xgtest = xgb.DMatrix(test, feature_names=feature)
    
watchlist = [ (xgtrain,'train'), (xgval, 'valid') ]
model = xgb.train(para, xgtrain, num_boost_round=3000, evals= watchlist, early_stopping_rounds=20, verbose_eval = 10)   
```

```python
#[713]	train-rmse:0.895479	valid-rmse:0.982936
```

```python
pred_Val_y = model.predict(xgtest)
```

```python
sub=pd.DataFrame(pred_Val_y,columns=['xg'])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 16))
xgb.plot_importance(model, max_num_features=20, height=0.5, ax=ax)
```
